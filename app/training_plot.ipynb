```python
import os

import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
device = torch.device("cpu")
device
```

lr = 0.0001

batch_size = 256

```python
if os.path.exists("./app/model/saved_model/dahalanishID/best_alexnet_model_checkpoint.pt"):
    checkpoint = torch.load(
        "./app/model/saved_model/dahalanishID/best_alexnet_model_checkpoint.pt"
        , map_location = device
    )
    best_epoch = checkpoint['epoch']
```

```python
best_epoch
```

```python
if os.path.exists("./app/model/saved_model/dahalanishID/latest_alexnet_model_checkpoint.pt"):
    checkpoint = torch.load(
        "./app/model/saved_model/dahalanishID/latest_alexnet_model_checkpoint.pt"
        , map_location = device
    )
    check_point_epoch = checkpoint['epoch']
    LOSS = checkpoint['loss']
    theta_training_loss = LOSS['theta_training_loss']
    theta_training_mae = LOSS['theta_training_mae']
    theta_validation_loss = LOSS['theta_validation_loss']
    theta_validation_mae = LOSS['theta_validation_mae']

    phi_training_loss = LOSS['phi_training_loss']
    phi_training_mae = LOSS['phi_training_mae']
    phi_validation_loss = LOSS['phi_validation_loss']
    phi_validation_mae = LOSS['phi_validation_mae']
```

```python
check_point_epoch
```

```python
X = range(1, check_point_epoch+1)
```

```python
# plt.subplot(2, 2, 1)
fig, ax = plt.subplots(2,2,figsize=(12,8))
ax[0, 0].plot(X, theta_training_loss, label="train theta")
ax[0, 0].plot(X, theta_validation_loss, label="val theta")
ax[0, 0].scatter(best_epoch, theta_validation_loss[best_epoch-1], color="red")
ax[0, 0].scatter(best_epoch, theta_training_loss[best_epoch-1], color="red")
ax[0, 0].set_xlabel("epoch")
ax[0, 0].set_ylabel("MSE")
# ax[0, 0].set_title("Epoch vs MSE Loss for theta")
ax[0, 0].legend()

ax[0, 1].plot(X, phi_training_loss, label="train phi")
ax[0, 1].plot(X, phi_validation_loss, label="val phi")
ax[0, 1].scatter(best_epoch, phi_validation_loss[best_epoch-1], color="red")
ax[0, 1].scatter(best_epoch, phi_training_loss[best_epoch-1], color="red")

ax[0, 1].set_xlabel("epoch")
ax[0, 1].set_ylabel("MSE")
# ax[0, 1].set_title("Epoch vs MSE Loss for phi")
ax[0, 1].legend()


# plt.subplot(2, 2, 1)
ax[1, 0].plot(X, theta_training_mae, label="train theta")
ax[1, 0].plot(X, theta_validation_mae, label="val theta")
ax[1, 0].scatter(best_epoch, theta_validation_mae[best_epoch-1], color="red")
ax[1, 0].scatter(best_epoch, theta_training_mae[best_epoch-1], color="red")
ax[1, 0].set_xlabel("epoch")
ax[1, 0].set_ylabel("MAE")
# ax[1, 0].set_title("Epoch vs MAE Loss for theta")
ax[1, 0].legend()

ax[1, 1].plot(X, phi_training_mae, label="train phi")
ax[1, 1].plot(X, phi_validation_mae, label="val phi")
ax[1, 1].scatter(best_epoch, phi_validation_mae[best_epoch-1], color="red")
ax[1, 1].scatter(best_epoch, phi_training_mae[best_epoch-1], color="red")

ax[1, 1].set_xlabel("epoch")
ax[1, 1].set_ylabel("MAE")
# ax[1, 1].set_title("Epoch vs MAE Loss for phi")
ax[1, 1].legend()
plt.show()
```

```python
plt.plot(X, phi_training_loss, label="train")
plt.plot(X, phi_validation_loss, label="val")
plt.scatter(best_epoch, phi_validation_loss[best_epoch-1], color="red")
plt.scatter(best_epoch, phi_training_loss[best_epoch-1], color="red")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Epoch vs MSE Loss for phi")
plt.legend()
plt.show()
```
